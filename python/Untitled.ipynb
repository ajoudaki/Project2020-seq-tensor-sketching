```python
%load_ext autoreload
%autoreload 2
import homology
from homology import *
run()
dists = homology.dists
```

```python
# MINIMIZER SKETCHING METHOD
edit_dists, types, sketch_dists = homology.ROC_curve(
    dists, False, sketch_exons=sketch_exons_only, matching_subsequences=matching_subsequences,
)
```

```python
fedists, fedists_str, ftypes, fsdists = [], [], [], []
for ed, t, sd in zip(edit_dists, types, sketch_dists):
    if 0 < ed and ed <= 15:
        fedists_str.append(f'{ed:02}')
        fedists.append(ed)
        ftypes.append(t)
        fsdists.append(sd)
sns.displot(x=fsdists, hue=fedists_str, hue_order=sorted(set(fedists_str)), kind='kde', palette='rocket', common_norm=False)
sns.displot(x=fsdists, hue=fedists_str, hue_order=sorted(set(fedists_str)), kind='kde', palette='rocket', common_norm=True)
```

```python
fedists, fedists_str, ftypes, fsdists = [], [], [], []
for ed, t, sd in zip(edit_dists, types, sketch_dists):
    if 0 < ed and ed <= 5:
        fedists_str.append(f'{ed:02}')
        fedists.append(ed)
        ftypes.append(t)
        fsdists.append(sd)
sns.displot(x=fsdists, hue=fedists_str, hue_order=sorted(set(fedists_str)), col=ftypes, kind='kde', palette='rocket')
```
